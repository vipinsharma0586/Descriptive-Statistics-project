```python
#Import the relevant python packages (numpy, pandas, matplotlib).

%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Get the mean of each row (car) and store it in variable 'mtcarsMeanCarWise'.

mtcarsMeanCarWise = mtcars.mean(axis = 1)
mtcarsMeanCarWise
```

```python
# Plot the density curves for each column of mtcars using 'for' loop and plot command.

%matplotlib inline
import matplotlib.pyplot as plt
mtcars.hist(bins=50, figsize=(20,15))
plt.show()

for cols in mtcars.columns:
    if (cols != "name"):
        mtcars[cols].plot(kind="density",figsize=(2,2))
        plt.title(cols)
        plt.show()

from pandas.plotting import scatter_matrix
attributes=["mpg","disp","hp", "wt", "qsec"]
scatter_matrix(mtcars[attributes],figsize=(12,8))
```

```python
# We use skewness as a measure of symmetry. If the skewness of is zero then the distribution represented is 
# perfectly symmetric. If the skewness is negative, then the distribution is skewed to the left, 
# while if the skew is positive then the distribution is skewed to the right. Hence, 
# it can also be considered as a measure of normality since normal distributions are symmetric and hence have a zero skew.
# Calculate the skeweness of each column of 'mtcars' using mtcars.skew()

mtcars.skew()
```

```python
# Extract cars with automatic transmission and cars with manual transmission 
# into separate variables 'mtCarsAutomatic' and 'mtCarsManual' respectively.

Automatic = mtcars.query('am == 0')
Manual = mtcars.query ('am == 1')
print(Automatic)
print(Manual)
```

```python
mtCarsAutomatic = Automatic['name']
mtCarsManual = Manual['name']
print(mtCarsAutomatic)
print(mtCarsManual)
```

```python
# Plot the box plot for 'mpg' (miles per gallons) for each of the subsets of cars grouped by transmission (automatic/ manual).

#1 - Automatic :

plt.boxplot(Automatic.mpg, showmeans = True, meanline = True)
plt.text(0.7,10.4,'min = 10.40')
plt.text(0.7,14.95,'25% = 14.95')
plt.text(1.15,17.30,'mean = 17.14')
plt.text(0.55,17.14,'50% (median) = 17.30')
plt.text(0.7,19.20,'75% = 19.20')
plt.text(0.7,24,'max = 24.40')
plt.title('Automatic Transmission')
```

```python
#2 - Manual :

df2 = pd.DataFrame(Manual['mpg'].describe())
df2
```
